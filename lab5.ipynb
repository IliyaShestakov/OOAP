```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_03.
#
# Выполнил: Шестаков И.А.
# Группа: АДЭУ_211

Ln = [] 
Ln1 = []

def is_lucky(num):
    ch = 0
    nech = 0
    for n in map(int, str(num)):
        if n % 2 == 0:
            ch += 1
        else:
            nech += 1
    if ch == nech:
        Ln.append(num)
    else:
        pass

def lucky_numbers(a, b):
    num = a   
    while num in range(a, b + 1):
        is_lucky(num)
        if num in Ln:
            Ln1.append(num)
            num += 1
        else:
            num += 1
    for i in Ln1:
        print(i, end=" ") 

a = int(input("Первый номер билета: "))
b = int(input("Последний номер билета: "))

print(lucky_numbers(a, b))

# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_04.
#
# Выполнил: Шестаков И.А.
# Группа: АДЭУ_211

def is_leap(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        A = True
    else:
        A = False
    return A

def days(month, year):
    dm31 = [1,3,5,7,8,10,12]
    dm30 = [4,6,9,11]
    if is_leap(year) == True:
        dm29 = [2]
    else:
        dm28 = [2]
    if month in dm31:
        date = 31
    elif month in dm30:
        date = 30
    elif month in dm29:
        date = 29
    else:
        date = 28
    return date

def next_date(day, month, year):
    if day == days(month, year):
        month += 1
        if month == 13:
            month = 1
            year += 1
            day = 1
        else:
            month += 1
            year = year
            day = 1
    else:
        day += 1
        month = month
        year = year
    return day, month, year

def previous_date(day, month, year):
    if day == 1:
        month -= 1
        if month == 0:
            year -= 1
            month = 12
            day = days(12, year)
        else:
            year = year
            month = month
            day = days(month, year)
    else:
        day -= 1
        month = month
        year = year
    return day, month, year

day, month, year = input('День, месяц, год через пробел: ').split()
day = int(day)
month = int(month)
year = int(year)
print('Предыдущий день: ',previous_date(day, month, year))
print('Следующий день: ',next_date(day, month, year))

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 3 2000
# Предыдущий день: 29/02/2000
# Следующий день: 02/03/2000
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_05.
#
# Выполнил: Шестаков И.А.
# Группа: АДЭУ_211

def is_leap(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        A = True
    else:
        A = False
    return A

def days(month, year):
    dm31 = [1,3,5,7,8,10,12]
    dm30 = [4,6,9,11]
    if is_leap(year) == True:
        dm29 = [2]
    else:
        dm28 = [2]
    if month in dm31:
        date = 31
    elif month in dm30:
        date = 30
    elif month in dm29:
        date = 29
    else:
        date = 28
    return date

def another_date(day, month, year, delta=1):
    if delta < 0:
        day += delta
        if day <= 0:
            month -= 1
            if month == 0:
                month = 12
                year -= 1
                day = days(12, year) + day
            else:
                month = month
                year = year
                day = days(month, year) + day
        else:
            day = day
            month = month
            year = year
        return day, month, year
    elif delta > 0:
        day += delta
        if day > days(month, year):
            day = day - days(month, year)
            month += 1
            if month > 12:
                month -= 12
                year += 1
            else:
                month += 1
                year = year
        else:
            day = day
            month = month
            year = year
        return day, month, year
    else:
        return day, month, year

day, month, year = input('День, месяц, год через пробел: ').split()
day = int(day)
month = int(month)
year = int(year)
delta = int(input('Свдиг (может быть отрицательным): '))
print('Новый день: ', another_date(day, month, year, delta))

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): -2
# Новый день: 30/12/1999
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): 2
# Новый день: 03/01/2000
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_14.
#
# Выполнил: Шестаков И.А.
# Группа: АДЭУ_211

OT = [] # список отрицательных
PO = [] # список положительных

numbers = list(map(int, input().split()))

def split_numbers(numbers):
    for n in numbers:
        if n < 0:
            OT.append(n)
        else:
            PO.append(n)
    PO.sort()
    OT.sort()
    OT.reverse()
    print(OT, ', ', PO, sep='')

print(split_numbers(numbers))   
# числа для примера: 1 4 -5 0 -33

# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_16.
#
# Выполнил: Шестаков И.А.
# Группа: АДЭУ_211


def pow1(value, power): # обычное возведение в степень
    v = value
    p = power
    return v ** p

def pow2(value, power): # рекурсивное возведение в степень
    if power == 1:
        return value
    elif power == 0:
        return 1
    else:
        return (value * pow2(value, power - 1))

print(pow1(5, 3))
print(pow2(5, 3))

# --------------
# Пример вывода:
#
# 125
# 125
```
