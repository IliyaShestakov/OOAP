```python
class Person: # (делали на паре 01.10)
    def __init__(self, n, s, q = 1): # конструктор класса
        self.name = n # Имя
        self.surname = s # Фамилия
        self.skill = q # Квалификация (по умолчанию = 1)
# Метод, который возвращает строку, включающую в себя всю информацию о сотруднике:
    def info(self):
        return "{0} {1}, {2}".format(self.name, self.surname, self.skill)
# Деструктор, который выводит на экрна фразу "До свидания мистер..."
    def __del__(self):
        print("Досвидания, мистер", self.name, self.surname)

worker = Person("И", "Котов", 3) # 1-ый объект класса Person
helper = Person("Д", "Мышев", 1) # 2-ой объект класса Person
maker = Person("О", "Рисов", 2) # 3-ий объект класса Person

print(worker.info())
print(helper.info())
print(maker.info())

del helper

try:
    input("Введите enter: ")
except SyntaxError:
    exit()
```

```python
class Student: 
    def __init__(self, n, s, p, u, g = "АДЭУ-211", vuz = "МГПУ"):
        self.s = s # Фамилия
        self.g = g # Группа, задается по умолчанию
        self.vuz = vuz # Т.к. группа задается по умолчанию, то и вуз тоже (логика)
        self.n = n # Имя 
        self.p = p # Пол
        self.u = u # успеваемость

    def __del__(self):
        print("Отчислен студент", self.n, self.s, d[self.u])

    def display(self):
        print("Имя: {}, фамилия: {}, ВУЗ: {}, группа: {}, пол: {}, успеваемость: {}".format(self.n, self.s, self.vuz, self.g, self.p, d[self.u]))

d = {2:'Задолжник', 4:'Ударник', 5:'Отличник'}

S1 = Student("Илья", "Шестаков", "М", 5)
S1.display()

S2 = Student("чел", "чел", "М", 2)
S2.display()

del S2

s1 = Student("Убили", "негра", "убили", 4)
s2 = Student("негра", "убили", "Ни за что ни про что, суки, замочили", 4)

if s1.u < s2.u:
  del s1
elif s1.u > s2.u:
  del s2
else:
  print('МОЛОДЦЫ!')
```
