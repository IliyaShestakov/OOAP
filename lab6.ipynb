```python
#      (Python).
# https://www.yuripetrov.ru/edu/python
#  task_06_02_01.
#
# :  ..
# : -211


def foo(a):
    """   , ,
           

    : O(N^2).
    """
    for i in range(len(a), 0, -1): # O(N)
        for j in range(1, i): # O(N)
            if a[j-1] > a[j]: # O(1)
                a[j-1], a[j] = a[j], a[j-1]
    return a


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python
#      (Python).
# https://www.yuripetrov.ru/edu/python
#  task_06_02_02.
#
# :  ..
# : -211


def foo(i):
    """    :
      = 0  0
      <0   
      >0    

    :
        - i (int): .

    : O(N).
    """
    digits = "0123456789"
    if i == 0: # O(1)
        return "0"
    result = ""
    while i > 0: # O(N)
        result = digits[i%10] + result
        i = i // 10
    return result

"""
i = int(-833)
foo(i)

i = int(0)
foo(i)
"""
i = int(55)
foo(i)
```

```python
#      (Python).
# https://www.yuripetrov.ru/edu/python
#  task_06_02_03.
#
# :  ..
# : -211


def foo(s):
    """       

    :
        - s (str): .

    : O(N).
    """
    val = 0
    for c in s: # O(N)
        if c.isdigit(): # O(1)
            val += int(c)
    return val

s = "7, 5, 8, 25"
foo(s)
```

```python
#      (Python).
# https://www.yuripetrov.ru/edu/python
#  task_06_02_04.
#
# :  ..
# : -211


def foo(n):
    """        

    :
        - n (int): .

    : # O(N^2).
    """
    res = []
    for i in range(1, n + 1): # O(N)
        divisors = 0
        j = 2
        while j < i and divisors == 0: # O(N)
            if i % j == 0: # O(1)
                divisors += 1
            j += 1

        if divisors == 0: # O(1)
            res.append(i)

    return res

n = int(19)
foo(n)
```
