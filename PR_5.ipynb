```python
class Win_Door:
    def __init__(self, x, y):
        self.square = x * y

class Room:
    def __init__(self, height, width, length, wd):
        self.square = 2 * height * (width + length)
        self.wd = []

    def addWD(self, w, h):
        self.wd.append(Win_Door(w, h))

    def workSurface(self):
        new_square = self.square
        for i in self.wd:
            new_square -= i.square
        return new_square

    def roolons(self, dr, wr):
        dr = int(input("Длина рулона: "))
        wr = int(input("Ширина рулона: "))
        roolon = dr * wr
        new_square = self.square
        for i in self.wd:
            new_square -= i.square
        needed = (new_square / roolon)
        return ('Понадобится {} рулонов'.format(needed))

r1 = Room(4, 5, 3, 0)
r1.addWD(1, 1) # пл. окна
r1.addWD(1, 1) # пл. окна
r1.addWD(1, 2) # пл. двери
print('Полная площадь стен (если в комнате не было ни окон ни дверей) =', r1.square)
print('Оклеиваемая площадь стен (с учетом окон и дверей) =', r1.workSurface())
r1.roolons(1, 1)
```
